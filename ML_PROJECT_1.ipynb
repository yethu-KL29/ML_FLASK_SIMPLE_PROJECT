```python
import pandas as pd
```

# Loading dataset

```python
data  = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
```

```python
data
```

## splitting training and test dataset

```python
X = data.drop('logS',axis=1)
```

```python
Y = data['logS']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size=0.2 , random_state=100)
```

```python
## Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
x_test
```

### importing linear regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
# training the model 
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(x_train , y_train)
```

```python
# apply the model to make prediction 
```

```python
train_pred = lr_model.predict(x_train)
test_pred  = lr_model.predict(x_test)
```

```python
len(train_pred)
```

```python
len(test_pred)
```

# Evaluating the prediction

```python

from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train,train_pred)
lr_train_r2 = r2_score(y_train, train_pred)

lr_test_mse = mean_squared_error(y_test,test_pred)
lr_test_r2 = r2_score(y_test,test_pred)
     
```

```python

print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```
